```python
import requests
import pandas as pd

APP_ID = "83700cd4"
API_KEY = "c97b9c32b474def2a4f57f418e8b2452"

url = "https://api.adzuna.com/v1/api/jobs/us/search/1"

params = {
    "app_id": APP_ID,
    "app_key": API_KEY,
    "results_per_page": 10,
    "what": "data analyst",
    "where": "New York"
}

response = requests.get(url, params=params)

data = response.json()

jobs = data['results']

job_list = []

for job in jobs:

    job_list.append({
        "Title": job.get("title"),
        "Company": job.get("company", {}).get("display_name"),
        "Location": job.get("location", {}).get("display_name"),
        "Category": job.get("category", {}).get("label"),
        "Salary_Min": job.get("salary_min"),
        "Salary_Max": job.get("salary_max"),
        "Created": job.get("created")
    })

df = pd.DataFrame(job_list)

df.head()
```

# =========================================================
# 4. BUILDING STRUCTURED DATASET
# =========================================================

```python
import requests
import pandas as pd
import time

APP_ID = "83700cd4"
API_KEY = "c97b9c32b474def2a4f57f418e8b2452"

roles = [
    "data analyst",
    "data scientist",
    "machine learning engineer",
    "ai engineer",
    "business analyst",
    "power bi developer"
]

all_jobs = []

for role in roles:

    for page in range(1, 6):

        url = f"https://api.adzuna.com/v1/api/jobs/us/search/{page}"

        params = {
            "app_id": APP_ID,
            "app_key": API_KEY,
            "results_per_page": 50,
            "what": role,
            "where": "New York"
        }

        response = requests.get(url, params=params)

        data = response.json()

        jobs = data.get("results", [])

        print(f"{role} - Page {page} - Jobs Found: {len(jobs)}")

        for job in jobs:

            all_jobs.append({
                "Role_Searched": role,
                "Title": job.get("title"),
                "Company": job.get("company", {}).get("display_name"),
                "Location": job.get("location", {}).get("display_name"),
                "Category": job.get("category", {}).get("label"),
                "Salary_Min": job.get("salary_min"),
                "Salary_Max": job.get("salary_max"),
                "Created": job.get("created"),
                "Contract_Type": job.get("contract_type"),
                "Description": job.get("description")
            })

        time.sleep(1)

df = pd.DataFrame(all_jobs)

print(df.shape)

df.head()
```

# 7. NLP-BASED SKILL EXTRACTION
# =========================================================

```python
df.groupby('Role_Searched')['Average_Salary'].mean()
```

```python
skills = [
    'python',
    'sql',
    'excel',
    'power bi',
    'tableau',
    'aws',
    'azure',
    'machine learning',
    'deep learning',
    'tensorflow',
    'pandas',
    'numpy',
    'spark',
    'hadoop',
    'statistics',
    'data visualization',
    'nlp',
    'artificial intelligence',
    'java',
    'c++'
]
```

```python
df['Description'] = df['Description'].astype(str)

df['Description_Lower'] = df['Description'].str.lower()
```

```python
skill_counts = {}

for skill in skills:

    count = df['Description_Lower'].str.contains(skill).sum()

    skill_counts[skill] = count

skill_counts
```

```python
skills_df = pd.DataFrame(
    skill_counts.items(),
    columns=['Skill', 'Demand_Count']
)

skills_df = skills_df.sort_values(
    by='Demand_Count',
    ascending=False
)

skills_df.head(10)
```

```python
import matplotlib.pyplot as plt

top_skills = skills_df.head(10)

plt.figure(figsize=(12,6))

plt.bar(top_skills['Skill'], top_skills['Demand_Count'])

plt.xticks(rotation=45)

plt.title("Top 10 Most Demanded Skills")

plt.xlabel("Skills")

plt.ylabel("Demand Count")

plt.show()
```

# =========================================================
# 12. BUSINESS SQL QUERIES & ANALYSIS
# =========================================================

```python
query = """
SELECT
    Company,
    COUNT(*) AS Total_Jobs
FROM jobs
GROUP BY Company
ORDER BY Total_Jobs DESC
LIMIT 10;
"""

top_companies_sql = pd.read_sql(query, conn)

top_companies_sql
```

```python
query = """
SELECT
    Role_Searched,
    ROUND(AVG(Average_Salary), 2) AS Avg_Salary
FROM jobs
GROUP BY Role_Searched
ORDER BY Avg_Salary DESC;
"""

salary_sql = pd.read_sql(query, conn)

salary_sql
```

```python
query = """
SELECT
    Location,
    COUNT(*) AS Total_Jobs
FROM jobs
GROUP BY Location
ORDER BY Total_Jobs DESC
LIMIT 10;
"""

location_sql = pd.read_sql(query, conn)

location_sql
```

```python
query = """
SELECT
    Contract_Type,
    COUNT(*) AS Total
FROM jobs
GROUP BY Contract_Type
ORDER BY Total DESC;
"""

contract_sql = pd.read_sql(query, conn)

contract_sql
```

```python
query = """
SELECT
    Company,
    ROUND(AVG(Average_Salary), 2) AS Avg_Salary
FROM jobs
GROUP BY Company
HAVING COUNT(*) >= 2
ORDER BY Avg_Salary DESC
LIMIT 10;
"""

company_salary_sql = pd.read_sql(query, conn)

company_salary_sql
```
